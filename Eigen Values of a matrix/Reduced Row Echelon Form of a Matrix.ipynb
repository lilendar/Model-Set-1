```python
# Code for eigen values and eigen vectors

from sympy import var, Eq, solve
import math


#TAKING THE INPUT FROM USER

a= int(input("No of Rows    "))
bb= int(input("No of Columns   "))
print("Enter the elements of the matrix in row ways ")
b=[]

for  i in range(0,a):
      c=[]
      for  j in range(0,bb):
         c.append(input())
      b.append(c) 

    
#PRINTING THE MATRIX

print("your matrix is :")    
print("1st 2nd & 3rd rows are : ", b)
print("\n")

ca=int(b[0][0])

cb=int(b[0][1])

cc=int(b[0][2])

cd=int(b[1][0])

cf=int(b[1][1])

cg=int(b[1][2])

ch=int(b[2][0])

cj=int(b[2][1])

ck=int(b[2][2])

coffconst=ca*((cf*ck)-(cg*cj))- cb*((cd*ck)-(ch*cg))+ cc*((cd*cj)-(ch*cf))
coffx=-ca*(cf+ck)-(cf*ck-(cg*cj))-cb*(-cd)+cc*ch
coffxsqr=(ca+(cf+ck))
x = var('x')
sol = solve(Eq(-x**3 + coffxsqr*x**2 + coffx*x + coffconst, 0), x)

#PRINTING THE EIGEN VALUE OF THE MATRIX

Eigenvalue1=sol[0]
Eigenvalue2=sol[1]
Eigenvalue3=sol[2]
print("Eigen values are",sol[0],sol[1],sol[2])
print("\n")


#SUBTRACT EIGEN VALUES FROM THE DIAGONAL ELEMENTS

diff1e1=ca-Eigenvalue1
diff2e1=cf-Eigenvalue1
diff3e1=ck-Eigenvalue1

diff1e2=ca-Eigenvalue2
diff2e2=cf-Eigenvalue2
diff3e2=ck-Eigenvalue2

diff1e3=ca-Eigenvalue3
diff2e3=cf-Eigenvalue3
diff3e3=ck-Eigenvalue3

cx=[[(diff1e1), cb, cc],[cd,(diff2e1),cg],[ch,cj,(diff3e1)]]

#PRINTING THE EIGEN VECTORS

Eigenvector1= [[(cb*cg)-(cc*diff2e1)], [-((diff1e1*cg)-(cc*cd))],[(diff1e1*diff2e1)-(cd*cb)]]
print("Eigen vectors for 1st eigen value is \n  ",Eigenvector1)

cx1=[[(diff1e2), cb, cc],[cd,(diff2e2),cg],[ch,cj,(diff3e2)]]
Eigenvector2= [[(cb*cg)-(cc*diff2e2)], [-((diff1e2*cg)-(cc*cd))],[(diff1e2*diff2e2)-(cd*cb)]]
cx2=[[(diff1e3), cb, cc],[cd,(diff2e3),cg],[ch,cj,(diff3e3)]]
print("Eigen vectors for 2nd eigen value is \n ",Eigenvector2)

Eigenvector3= [[(cb*cg)-(cc*diff2e3)], [-((diff1e3*cg)-(cc*cd))],[(diff1e3*diff2e3)-(cd*cb)]]
print("Eigen vectors for 3rd eigen value is \n ",Eigenvector3)

#DYNAMIC SYSTEM

Powmatrix=int(input("Enter the power/state of the matrix you entered for 1st eigen value \n   "))
powfactor=Eigenvalue1**Powmatrix
n1=powfactor*((cb*cg)-(cc*diff2e1))
n2=powfactor*-((diff1e1*cg)-(cc*cd))
n3=powfactor*((diff1e1*diff2e1)-(cd*cb))
new1=[n1,n2,n3]
print(" new state of the system \n",new1," \n system in neutral ")



Powmatrix2=int(input("Enter the power/state of the matrix you entered for 2nd eigen value \n  "))
powfactor1=Eigenvalue2**Powmatrix2
n4=powfactor1*((cb*cg)-(cc*diff2e2))
n5=powfactor1*-((diff1e2*cg)-(cc*cd))
n6=powfactor1*((diff1e2*diff2e2)-(cd*cb))
new2=[n4,n5,n6]
print(" new state of the system \n",new2,"\n system in unstable ")



Powmatrix3=int(input("Enter the power/state of the matrix you entered for 3rd eigen value \n"))
powfactor2=Eigenvalue3**Powmatrix3
n7=powfactor2*((cb*cg)-(cc*diff2e3))
n8=powfactor2*-((diff1e3*cg)-(cc*cd))
n9=powfactor2*((diff1e3*diff2e3)-(cd*cb))
new3=[n7,n8,n9]
print(" new state of the system \n",new3," \n system in unstable ")
```

```python
def ToReducedRowEchelonForm( M):
    if not M: return
    lead = 0
    rowCount = len(M)
    columnCount = len(M[0])
    for r in range(rowCount):
        if lead >= columnCount:
            return
        i = r
        while M[i][lead] == 0:
            i += 1
            if i == rowCount:
                i = r
                lead += 1
                if columnCount == lead:
                    return
        M[i],M[r] = M[r],M[i]
        lv = M[r][lead]
        M[r] = [ mrx / float(lv) for mrx in M[r]]
        for i in range(rowCount):
            if i != r:
                lv = M[i][lead]
                M[i] = [ iv - lv*rv for rv,iv in zip(M[r],M[i])]
        lead += 1
 
 
mtx = [
   [ 1, 2, -1,4],
   [ 2, 3, -1,3],
   [-2, 0, -3, 22],]
 
ToReducedRowEchelonForm( mtx )
 
for rw in mtx:
  print( ', '.join( (str(rv) for rv in rw) ))
```
